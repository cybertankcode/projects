```python
import pandas as pd
data = pd.read_csv(".\\datasets\\Cyber_Datasets\\AWS_Honeypot.csv")
data.info()
data.head()
```

```python
data['datetime'] = pd.to_datetime(data['datetime'])
print(data.head(5))
print(data.tail(5))
print(data.info())
```

```python
china = data[data['country']=="China"]
plt.plot(china['datetime'], china["dpt"])
plt.title("China:Cumulative Dpt")
plt.xlabel("Date")
plt.ylabel("Number of dpt")
plt.show()
```

```python
Germany = data[data['country']=="Germany"]
plt.plot(Germany['datetime'], Germany["dpt"])
plt.title("Germany:Cumulative dpt")
plt.xlabel("Date")
plt.ylabel("Number of dpt")
plt.show()
```

```python
South_Korea = data[data['country'] == 'South Korea']

plt.plot(South_Korea['datetime'], South_Korea['dpt'])
plt.title('South_Korea: Cumulative dpt')
plt.xlabel('Date')
plt.ylabel('Number of dpt')
plt.show()
```

```python
Tunisia = data[data['country'] == 'Tunisia']

plt.plot(Tunisia['datetime'], Tunisia['dpt'])
plt.title('Tunisia: Cumulative dpt')
plt.xlabel('Date')
plt.ylabel('Number of dpt')
plt.show()
```

```python
def plot_cumulative_cases(country_name):
    country = data[data['country'] == country_name]
    plt.plot(country['datetime'], country['dpt'])
    plt.title('{}: Cumulative Reported'.format(country_name))
    plt.xlabel('Date')
    plt.ylabel('Number of dpt')
    plt.show()
```

```python
plot_cumulative_cases("United States")
```

```python
plot_cumulative_cases("Tunisia")
```

```python
plt.plot(South_Korea['datetime'], South_Korea['dpt'], label="Poland")
plt.plot(Germany['datetime'], Germany["dpt"], label="Germany")
plt.plot(Tunisia['datetime'], Tunisia["dpt"], label="Tunisia")
plt.legend()
plt.show()
```
