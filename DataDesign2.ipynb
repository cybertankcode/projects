```python
sp500 = pd.read_csv(".\\datasets\\Other_Datasets\\sp500_companies\\sp500_index.csv")
date = sp500["Date"]
stocks = sp500["S&P500"]

date = pd.to_datetime(date)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1, figsize=(7,8))

axes = [ax1, ax2, ax3, ax4]
for ax in axes:
    ax.plot(date , stocks, color='#af0b1e', alpha=0.1)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.tick_params(top=False, left=False)
    for location in ['left', 'right', 'top', 'bottom']:
        ax.spines[location].set_visible(False)
        
ax1.plot(date[:2012], stocks[:2012],
         color='#af0b1e', linewidth=2.5)
ax2.plot(date[2011:2015], stocks[2011:2015],
         color='#af0b1e', linewidth=2.5)
ax3.plot(date[2014:2016], stocks[2014:2016],
         color='#af0b1e', linewidth=2.5)
ax4.plot(date[2015:], stocks[2015:],
         color='#af0b1e', linewidth=2.5)

ax1.text(15500, 1900, '88', alpha=0.5)
ax1.text(16500, 2300, '200', alpha=0.5)
ax1.text(17500, 3000, '500', alpha=0.5)

ax1.text(15500, 900, '2012-2015', color='#af0b1e',
         weight='bold', rotation=3)

ax2.text(16500, 800, '2015-2018', color='#af0b1e', weight='bold')
ax3.text(17500, 500, '2018-2019', color='#af0b1e', weight='bold')
ax4.text(18500, 600, '2019-2021', color='#af0b1e', weight='bold',rotation=45)

ax1.text(16500, 5000, 'Year vs Stocks',
         size=14, weight='bold')
ax1.text(16200, 4500, 'Chart showing stocks per year',
         size=12) 

plt.show()
```

```python
sp500 = pd.read_csv(".\\datasets\\Other_Datasets\\sp500_companies\\sp500_index.csv")
date = sp500["Date"]
stocks = sp500["S&P500"]

date = pd.to_datetime(date)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1, figsize=(7,8))

axes = [ax1, ax2, ax3, ax4]
for ax in axes:
    ax.plot(date , stocks, color='#af0b1e', alpha=0.1)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.tick_params(top=False, left=False)
    for location in ['left', 'right', 'top', 'bottom']:
        ax.spines[location].set_visible(False)
        
ax1.plot(date[:2012], stocks[:2012],
         color='#af0b1e', linewidth=2.5)
ax2.plot(date[2011:2015], stocks[2011:2015],
         color='#af0b1e', linewidth=2.5)
ax3.plot(date[2014:2016], stocks[2014:2016],
         color='#af0b1e', linewidth=2.5)
ax4.plot(date[2015:], stocks[2015:],
         color='#af0b1e', linewidth=2.5)

ax1.text(15500, 1900, '88', alpha=0.5)
ax1.text(16500, 2300, '200', alpha=0.5)
ax1.text(17500, 3000, '500', alpha=0.5)

ax1.text(15500, 900, '2012-2015', color='#af0b1e',
         weight='bold', rotation=3)

ax2.text(16500, 800, '2015-2018', color='#af0b1e', weight='bold')
ax3.text(17500, 500, '2018-2019', color='#af0b1e', weight='bold')
ax4.text(18500, 600, '2019-2021', color='#af0b1e', weight='bold',rotation=45)

ax1.text(16500, 5000, 'Year vs Stocks',
         size=14, weight='bold')
ax1.text(16200, 4500, 'Chart showing stocks per year',
         size=12) 


stocks = [200, 500, 1200, 1601 ]
proportions = [round(stock/1601, 4) for stock in stocks]

xmax_vals = [round(0.5 + proportion * 0.3, 3) for proportion in proportions]
print(proportions)
print(xmax_vals)

for ax, xmax, stocks in zip(axes, xmax_vals, stocks):
    # First line
    ax.axhline(y=1600, xmin=0.5, xmax=0.8,
               linewidth=6, color='#af0b1e', alpha=0.1)

    # Second line
    ax.axhline(y=1600, xmin=0.5, xmax=xmax,
               linewidth=6, color='#af0b1e')
    
    ax.text(17500, 1850, format(stocks, ','),
            color='#af0b1e', weight='bold')

for ax in axes:
    ax.axhline(y=1200, xmin=0.3, xmax=0.9, linewidth=6, color='#af0b1e', alpha=0.1)

plt.show()
```
