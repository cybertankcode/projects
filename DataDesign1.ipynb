```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sp500 = pd.read_csv(".\\datasets\\Other_Datasets\\sp500_companies\\sp500_index.csv")
sp500.info()
sp500.head()
```

```python
date = sp500["Date"]
stocks = sp500["S&P500"]
print(date)
print(stocks)
```

```python
plt.plot(date, stocks)
plt.ticklabel_format(axis='y', style="plain")
plt.title("Stocks vs datetime")
plt.xlabel("Datetime")
plt.ylabel("Stocks")
plt.show()
```

```python
plt.barh(date,stocks)
plt.title("Stocks vs datetime")
plt.xlabel("Datetime")
plt.ylabel("Stocks")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(sp500['Date'],
         sp500['S&P500'],
       height=0.45, color='#af0b1e')

for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)
    
ax.xaxis.tick_top()
ax.tick_params(top=False, left=False)
ax.tick_params(axis='x', colors='grey')
ax.set_xticklabels(['0', '2,000', '3,000'])
ax.text(x=0.5, y=3100,
        s='Dates vs stocks',
        weight='bold', size=17)
ax.text(x=0.5, y=3000.5,
        s='Top 500 Stocks)',
        size=12)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(sp500['Date'],
         sp500['S&P500'],
       height=0.45, color='#af0b1e')

for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)
    
ax.xaxis.tick_top()
ax.tick_params(top=False, left=False)
ax.tick_params(axis='x', colors='grey')
ax.set_xticklabels(['0', '2,000', '3,000'])
ax.text(x=0.5, y=3100,
        s='Dates vs stocks',
        weight='bold', size=17)
ax.text(x=0.5, y=3000.5,
        s='Top 500 Stocks)',
        size=12)

ax.set_yticklabels([]) # an empty list removes the labels
date = sp500['Date']
for i, datetime in zip(range(0,20), date):
    ax.text(x=-1000, y=i-0.15, s=datetime)

ax.axvline(x=2000, ymin=0.045, c='grey', alpha=0.5)

plt.show()
```
